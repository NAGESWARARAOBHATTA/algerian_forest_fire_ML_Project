```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("Algerian_forest_fires_dataset.csv")
df.head()
```

```python
df.drop(df.index[122],inplace=True)
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,"region"]=0
df.loc[122:,"region"]=1
```

```python
df.info()
```

```python
df[["region"]]=df[["region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df[['month','year','Temperature', 'RH', 'Ws']]=df[['month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
obj=[features for features in df.columns if df[features].dtype=='O']
obj
```

```python
for i in obj:
    if i !='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv('Algerian_forest_fire_cleanned_dataset.csv',index=False)
```
